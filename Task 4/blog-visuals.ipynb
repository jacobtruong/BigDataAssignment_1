```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from cassandrautils import *
```

```python
weather = getWeatherDF()

weather['forecast_timestamp'] = pd.to_datetime(weather['forecastdate'], format="%d-%m-%Y %H:%M")
print("Weather from ", weather['forecast_timestamp'].min(), " To ", weather['forecast_timestamp'].max())

weather
```

```python
london = weather[weather.location == "London"]
london_dates = pd.to_datetime(london['forecastdate'], format="%d-%m-%Y %H:%M")
london_dates
```

```python
tokyo = weather[weather.location == "Tokyo"]
tokyo_dates = pd.to_datetime(tokyo['forecastdate'], format="%d-%m-%Y %H:%M")
tokyo_dates
```

```python
plt.figure(figsize=(20,6))
plt.plot(london_dates, london['temp'])
plt.xlabel("Time")
plt.ylabel("Temperature")
plt.title("Total London weather data recorded")
```

```python
plt.figure(figsize=(20,6))
plt.plot(tokyo_dates, tokyo['temp'])
plt.xlabel("Time")
plt.ylabel("Temperature")
plt.title("Total Tokyo weather data recorded")
```

```python
mimesis = getMimesisDF()
mimesis
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

ax.set_title('Education level below the age of 25')


mimesis_below_25 = mimesis[mimesis.age <= 25]
bachelor_below_25 = len(mimesis_below_25[mimesis_below_25.education == "Bachelor"])
master_below_25 = len(mimesis_below_25[mimesis_below_25.education == "Master"])
phd_below_25 = len(mimesis_below_25[mimesis_below_25.education == "PhD"])

print(len(mimesis_below_25))

pie_data = np.array([bachelor_below_25, master_below_25, phd_below_25])
pie_labels = ["Bachelor", "Master", "PhD"]

ax.pie(pie_data, labels = pie_labels,  startangle = 90, autopct='%1.0f%%')
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

ax.set_title('Education level between the age of 26 and 50')


mimesis_tmp = mimesis[mimesis.age > 25]
mimesis_26_50 = mimesis_tmp[mimesis_tmp.age <= 50]
bachelor_26_50 = len(mimesis_26_50[mimesis_26_50.education == "Bachelor"])
master_26_50 = len(mimesis_26_50[mimesis_26_50.education == "Master"])
phd_26_50 = len(mimesis_26_50[mimesis_26_50.education == "PhD"])

print(len(mimesis_26_50))

pie_data = np.array([bachelor_26_50, master_26_50, phd_26_50])
pie_labels = ["Bachelor", "Master", "PhD"]

ax.pie(pie_data, labels = pie_labels,  startangle = 90, autopct='%1.0f%%')
```
